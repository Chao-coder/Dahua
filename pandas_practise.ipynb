```python


# 用Series 字典对象生成DataFrame
df2 = pd.DataFrame({'A':1.,
                    'B':pd.Timestamp('20230606'),
                    'C':pd.Series(2,index = list(range(6)),dtype = 'float32'),
                    'D':np.array([5]*6,dtype='int32'),
                    'E':pd.Categorical(['test','train','РЖД','рынок','Опять','забрать']),
                    'F':'wcc'
})
df2
```
